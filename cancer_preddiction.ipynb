```python
df["GENDER"]=df.GENDER.apply(lambda x: 1 if x=="M" else 0)
```

```python
df["LUNG_CANCER"]=df.LUNG_CANCER.apply(lambda x: 1 if x=="YES" else 0)
```

```python
pd.crosstab(df.GENDER,df.LUNG_CANCER)
```

```python
pd.crosstab(df.GENDER,df.LUNG_CANCER).plot(kind="bar",figsize=(10,6),color=["salmon","lightblue"])
plt.title("GENDER vs LUNG_CANCER")
plt.xlabel("GENDER")
plt.ylabel("LUNG_CANCER")
plt.legend(["no disease","disease"])
plt.xticks(rotation=0)
plt.show()
```

```python
sns.distplot(df['LUNG_CANCER'],kde=True)
```

```python
plt.figure(figsize=(12,6))

plt.scatter(
    df.AGE[df.LUNG_CANCER == 1],
    df.SMOKING[df.LUNG_CANCER ==1],
    c="salmon"
)

plt.scatter(
    df.AGE[df.LUNG_CANCER == 0],
    df.SMOKING[df.LUNG_CANCER == 0],
    c="black"
)

plt.legend(["disease", "no disease"])
plt.xlabel("Age")
plt.ylabel("Smoking Level")
plt.title("Age vs Smoking by Lung Cancer Status")

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm",linewidths=0.5)
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
X = df.drop("LUNG_CANCER", axis=1)
y = df["LUNG_CANCER"]

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print(y_resampled.value_counts())
```

```python
X.isnull().sum()
```

```python
X.shape,y.shape,df.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
LGModel=LogisticRegression()
```

```python
LGModel.fit(X_train,y_train)
```

```python
LGModel_y_pred=LGModel.predict(X_test)
```

```python
print(classification_report(y_test,LGModel_y_pred))
```

```python
RFModel=RandomForestClassifier()
```

```python
RFModel.fit(X_train,y_train)
```

```python
RFModel_y_pred=RFModel.predict(X_test)
```

```python
print(classification_report(y_test,RFModel_y_pred))
```

```python
GDBModel=GradientBoostingClassifier()
```

```python
GDBModel.fit(X_train,y_train)
```

```python
GDBModel_y_pred=GDBModel.predict(X_test)
```

```python
print(classification_report(y_test,GDBModel_y_pred))
```
